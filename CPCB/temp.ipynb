```python
import pandas as pd

x_df=pd.read_csv(r"/home/rishi/ML Projects/Air Pollution/sites_combv2/site_134_Police_Commissionerate_Jaipur_RSPCB_15Min.csv")

x_df['Timestamp'] = pd.to_datetime(x_df['Timestamp'], errors='coerce')
        
# Drop rows with invalid timestamps
x_df = x_df.dropna(subset=['Timestamp'])

# Set as index (required for resample)
x_df = x_df.set_index('Timestamp')

# Sort by timestamp
x_df = x_df.sort_index()

# Resample to hourly using median
x_df = x_df.resample('1h').median()

# Reset index to make Timestamp a column again (for CSV saving)
x_df = x_df.reset_index()
x_df
```
