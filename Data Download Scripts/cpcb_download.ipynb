```python
s=""
import base64, json

decoded = json.loads(base64.b64decode(s).decode())
decoded
x=decoded['dropdown']['stations']
x
u_list=[]
for city, stations in x.items():
    for site in stations:
        value=site['value']
        label=site['label']
        m=value+' ' +label
        m=m.replace(',', ' ').replace('-', '').replace('  ',' ').replace(' ', '_')+"_15Min.csv"
        u_list.append(m)
u_list


flist=[]
for year in range(2010,2026):
    for site in u_list:
        u=f"https://airquality.cpcb.gov.in/dataRepository/download_file?file_name=Raw_data/15Min/{year}/{site}"
        flist.append(u)
flist
with open('cpcb_urls_3.txt','w') as f:
    f.write('\n'.join(flist))
import requests
from concurrent.futures import ThreadPoolExecutor
import os
import itertools
import time

base = "https://airquality.cpcb.gov.in"
os.makedirs("cpcb_data3", exist_ok=True)


def create_session():
    s = requests.Session()
    s.get(base + "/ccr/")
    xsrf = s.cookies.get("_xsrf")

    s.headers.update({
        "Referer": base + "/ccr/",
        "Origin": base,
        "User-Agent": "Mozilla/5.0",
        "X-XSRFToken": xsrf,
    })
    return s


SESSION_POOL = [create_session() for _ in range(6)]
session_cycle = itertools.cycle(SESSION_POOL)


def valid_response(r, first_chunk):
    if r.status_code != 200:
        return False
    if "text/html" in r.headers.get("Content-Type", ""):
        return False
    if b"Internal Server Error" in first_chunk:
        return False
    return True


def download(url):
    try:
        s = next(session_cycle)  # reuse session

        r = s.get(url, stream=True, timeout=60)

        file_part = url.split("file_name=")[1]
        parts = file_part.split("/")
        year = parts[2]
        orig_name = parts[-1]

        fname = f"{year}_{orig_name}"
        path = os.path.join("cpcb_data4", fname)

        first_chunk = next(r.iter_content(8192))

        if not valid_response(r, first_chunk):
            print(" Skipping:", fname)
            return

        print("", fname)

        with open(path, "wb") as f:
            f.write(first_chunk)
            for chunk in r.iter_content(8192):
                f.write(chunk)

        time.sleep(0.2)  # lets OS recycle sockets

    except Exception as e:
        print(" Error:", e)


urls = open("cpcb_urls_3.txt").read().splitlines()

# keep threads moderate
with ThreadPoolExecutor(10) as ex:
    ex.map(download, urls)
```
