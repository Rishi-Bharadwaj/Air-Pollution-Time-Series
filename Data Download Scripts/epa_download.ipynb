```python
base="https://aqs.epa.gov/aqsweb/airdata/hourly_44201_1984.zip"
urls=[]
years=range(2025,2026)
codes=[44201,42401,42101,42602,88101,81102, "WIND","TEMP" ,"PRESS", "RH_DP"]
```

```python
for year in years:
    for code in codes:
        urls.append(f"https://aqs.epa.gov/aqsweb/airdata/hourly_{code}_{year}.zip")
urls
```

```python
with open("urls.txt", "w") as f:
    f.write("\n".join(urls))
```
