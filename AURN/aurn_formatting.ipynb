```python
def rdata_to_dataframe(file_path):
    parsed = rdata.parser.parse_file(fr"{file_path}")
    converted = rdata.conversion.convert(parsed)
    col_name=list(converted.keys())[0]
    df=pd.DataFrame(converted[col_name])
    df["date"] = pd.to_datetime(df["date"], unit="s", utc=True)
    return df
```

```python
folder = Path(r"/home/rishi/ML Projects/Air Pollution/Datasets/aurn_data")

files = [f.name for f in folder.iterdir() if f.is_file()]
print(files)
len(files)
files=sorted(files)
files
```

```python
base_path=r"/home/rishi/ML Projects/Air Pollution/Datasets/aurn_data"
df_list=[]
for file in tqdm(files):
    curr_file=os.path.join(base_path,file)
    curr_df=rdata_to_dataframe(file_path=curr_file)
    df_list.append(curr_df)
```

```python
big_df = pd.concat(df_list, ignore_index=True)
```
